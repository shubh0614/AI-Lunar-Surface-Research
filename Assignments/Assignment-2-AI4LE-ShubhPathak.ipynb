#### Q1 [3 points]

***Given a 2d numpy array, the task is to flatten a 2d numpy array into a 1d array.***

```
[[1, 2, 3], 
[2, 4, 5], 
[6, 8, 3]]
```

````python
## Soln 1
import numpy as np
# Create 2D array
a=np.array([[1, 2, 3], 
[2, 4, 5], 
[6, 8, 3]])

# Flatten it
flat_arr=a.reshape(-1)

# Print the output along with the dimension of flatten array
dim= flat_arr.ndim
print("Flattened Array : {} \nDimension of flattened array : {}".format(flat_arr, dim))
````

#### Q6 [10 Points]

***Create a whirlpool pattern using numpy arrays***  

**1) Understand what is a whirlpool -**
- *Whirpool starts from 0 at the centre of an array*
- *Every layer of whirlpool is incremented by 1*
- *The last layer of whirlpool can have any number but only between 1 to 10*

**2) Example to visualize the whirlpool whose last layer is filled with 3s -**
- *Whirpool starts from 0 at the center*
- *The first layer is filled with 1s*
- *The second layer is filled with 2s*
- *The third/last layer is filled with 3s*

**Below you can see the layers of whirlpool for the above example -**        
1.png

**Here is the reason why we named it a whirlpool pattern -**       
2.png

**3) Requirments of this question -**        
- *Create a function that generates such whirpool patterns given the digits to be used in last layer*
- *The last layer of whirpool should only allow numbers between 1 to 10 (inclusive).*
- *No hardcoding is allowed!* 


````python
## Soln 6
import numpy as np

# Function to create whirlpool patterns

def whirlpool(last_layer_element):
    no_of_elements=(last_layer_element*2)+1
    mat = np.zeros((no_of_elements, no_of_elements),dtype=np.int64)
    x, y = 0, 0
    for i in range(0,no_of_elements ):
        y = 0
        for j in range(0, no_of_elements ):
            if abs(i-last_layer_element)>abs(j-last_layer_element):
                mat[x][y]=abs(i-last_layer_element)
            else:
                mat[x][y]=abs(j - last_layer_element)
            y += 1
        x += 1
        
    print(mat)

# Call the function
last_layer_element = int(input("Please Enter last layer element  : "))
whirlpool(last_layer_element)

````


````python
## Soln 7 
import pandas as pd

# Convert the given data to dict
rawdata=input("Enter raw data : ")
data_dict={}
for i in rawdata.split(','):
    x=i.split(" - ")
    data_dict[x[0]]=int(x[1])

    
# Code the function

def high_scorer(data_dict):
    sort_choice=input("Enter t for top n values or b bottom n values :")
    n=int(input("Enter number of performers to be displayed : "))
    key_column_name=input("Enter Key column heading : ")
    value_column_name=input("Enter Value column heading : ")
    
    df=pd.DataFrame(data_dict.items(), columns=[key_column_name, value_column_name])
    if sort_choice=='t':
        df=df.sort_values(value_column_name, ascending=False)
        print("|       Top 3       |")
        print("---------------------")
    elif sort_choice=='b':
        df=df.sort_values(value_column_name)
        print("|       Bottom 3       |")
        print("------------------------")
    else:
        print("Wrong input!")
    
    df.insert(0,'Rank',range(1,len(data_dict)+1))
    df.set_index("Rank",inplace=True)
    display((df.head(n)))


# Call the function on dict data to fetch top 3 performers
high_scorer(data_dict)
````

***Upload this [image](https://drive.google.com/file/d/1KUv8KUmDcudp1t5e9xZBwvBHFREN-jND/view?usp=sharing/) to kaggle notebook and create a pencil sketch of it.***     
***The final output of the code should be two images; Original and Sketch as shown below:-***      
<img src = "https://drive.google.com/uc?id=1-oyVxR6NLbd16zw6zen-7bKKswRZRk1I">

````python
## Soln 10
import cv2 
import matplotlib.pyplot as plt

# Read the image and convert it into an array
img= cv2.imread("../input/image01/original_image.jpg")

# Get the grayscale image
gray_scale_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Invert the gray image
invert_img=cv2.bitwise_not(gray_scale_img)

# Blur the inverted image
blur_img=cv2.GaussianBlur(invert_img, (25,25), 0)

# Invert the blur image
invert_blur_img=cv2.bitwise_not(blur_img)

# Blend (divide) gray and inverted blur images - Pencil Sketch is ready
sketch=cv2.divide(gray_scale_img,invert_blur_img,scale=256.0)

# Display the original and Pencil sketch images using suplots
plt.figure(figsize=(15,8))
values={
    'img_label':["Original image","Pencil Sketch of the Original Image"] ,
    'colors':['red','k'] ,
    'image':[img,sketch] 
}

for i in range(0,2):
    plt.subplot(1,2,i+1)
    plt.title(values['img_label'][i],color=values['colors'][i])
    plt.imshow(values['image'][i],cmap="gray")
    plt.axis('off')

plt.show()
````
