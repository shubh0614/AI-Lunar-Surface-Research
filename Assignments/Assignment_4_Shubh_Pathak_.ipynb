***Visualise the process of convergence using different learning rates during a total of 100 iterations.***

Requirments of this question for each learning rate:-

- *How many iterations did the model take to converge?*       
- *If the model didn't converge then what is the value of x after 100 iterations?*        
- *Generate the subplots of the function with all the points in the process of convergence for each learning rate*

````python
## Soln - 7: Effect of learning rates on convergence
import  numpy as np
import matplotlib.pyplot as plt
# Write your code below

initial_x = 10
learning_rate = [0.01, 0.05, 0.1, 0.4, 0.5, 0.9, 0.99, 0.999, 1]

def grad_descent(learning_rate):
    l=learning_rate
    x = 10
    iteration=100
    x_val = []
    for a in range(iteration):
        y_derivative= 2*x-4
        dx=l*y_derivative
        x =x-dx

        if abs(dx) <= 1e-3:
            return x_val
        else:
          x_val.append(x)

    return x_val



X = np.arange(-12, 17)
y = X ** 2 - 4 * X + 1

for i in range(9):
  x_vals = np.array(grad_descent(learning_rate[i]))
  y_vals = x_vals ** 2 - 4 * x_vals + 1
  fig, axis = plt.subplots(figsize=(7,5))
  axis.set_facecolor("black")
  axis.plot(X, y)
  axis.plot(x_vals,y_vals,color='red',lw=2,linestyle="dashed",marker='o')
  axis.axvline(2, color ='red', lw = 1,linestyle="dashed", alpha = 0.75)
  axis.set_xlabel(f"Took {len(x_vals)} iterations to converge at {x_vals[-1]} at a learning rate of {learning_rate[i]}\n")

plt.show()




# Print your conclusions below
print("""
1. At lower learning rate the model takes less iterations to converge.
2. At very high learning rate the model takes more than 100 iteration to converge. 
3. At Learning rate= [0.99,0.999,1.0] the model doesn't converge in 100 iterations.
4. At Learning rate= [0.1,0.4,0.5,0.9] the model converges in less than 100 iterations.
5. At Learning rate= 0.5 the model converges at 2 in 1 iteration only.
6. Learning rate=[0.1,0.4] are good for the model as a desirable learning rate is one that's low enough so that the network converges to something useful but high enough so that it can be trained in lesstime and less resources and these learning rates are of reasonable value and converge at the desired output in less iteration and less time.
""")
````
